```python
from scipy.spatial.distance import cdist
import random

# Function to check if a new point satisfies the distance constraint
def satisfies_distance_constraint(new_point, selected_points, min_distance):
    distances = cdist([new_point], selected_points)[0]
    return all(distance > min_distance for distance in distances)

# Select 6 points with distances greater than 1.5 miles
min_distance = 1.5  # Minimum distance in miles
selected_points = []
num_points_to_select = 6

while len(selected_points) < num_points_to_select:
    new_point = random.choice(df['geometry'])
    if satisfies_distance_constraint(new_point, selected_points, min_distance):
        selected_points.append(new_point)

# Create a new GeoDataFrame with the selected points
selected_data = {'geometry': selected_points}
selected_gdf = gpd.GeoDataFrame(selected_data, crs="EPSG:4326")

print(selected_gdf)
```
